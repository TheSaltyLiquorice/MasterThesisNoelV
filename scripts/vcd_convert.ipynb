```python
#!/usr/bin/env python3
import json
import sys
from pyDigitalWaveTools.vcd.parser import VcdParser

fname = 'vcd_trace.vcd'
with open(fname) as vcd_file:
    vcd = VcdParser()
    vcd.parse(vcd_file)
    data = vcd.scope.toJson()
    #f = open('test.json','w')
    data = json.dumps(data, indent=4, sort_keys=True)
    x = json.loads(data)
for element in x['children'][0]['children'][0]['children'][0]['children']:
    print(element['name'].strip('\\')+' = '+str(element['type']['width'])+'\''+element['data'][0][1]+';')
```
